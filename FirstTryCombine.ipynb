```python
import pandas as pd
```

```python
df_jjj = pd.read_csv("jjj_data.csv", sep='\t')
df_jjj = df_jjj.reset_index()
df_jjj_rename = df_jjj.copy()
df_jjj_rename["title"] = df_jjj_rename["title"].str.replace(".","")
df_jjj_rename["title"] = df_jjj_rename["title"].str.replace("(","")
df_jjj_rename["title"] = df_jjj_rename["title"].str.replace(")","")
df_jjj_rename["title"] = df_jjj_rename["title"].str.replace("'","")
df_jjj_rename["title"] = df_jjj_rename["title"].str.replace(" ","")
df_jjj_rename["title"] = df_jjj_rename["title"].str.replace(",","")
df_jjj_rename["title"] = df_jjj_rename["title"].str.upper()

df_jjj_rename
```

```python
df_spotify = pd.read_csv("spotify_data.csv")
df_spotify.head()
```

```python
df_spotify = df_spotify.reset_index()
```

```python
df_spotify_rename = df_spotify.copy()
df_spotify_rename["name"] = df_spotify_rename["name"].str.replace(".","")
df_spotify_rename["name"] = df_spotify_rename["name"].str.replace("(","")
df_spotify_rename["name"] = df_spotify_rename["name"].str.replace(")","")
df_spotify_rename["name"] = df_spotify_rename["name"].str.replace("'","")
df_spotify_rename["name"] = df_spotify_rename["name"].str.replace(" ","")
df_spotify_rename["name"] = df_spotify_rename["name"].str.upper()

#df_spotify["artists"] = df_spotify["artists"].str.strip("[]")
#df_spotify["artists"] = df_spotify["artists"].str.strip("'")
#df_spotify["artists"] = df_spotify["artists"].str.replace("'","")
#df_spotify["artists"] = df_spotify["artists"].str.replace(",","")
#df_spotify["artists"] = df_spotify["artists"].str.replace(" ","")
df_spotify_rename
```

```python
df = df_jjj_rename.merge(df_spotify_rename, left_on=["title"], right_on=["name"], how="left")
index_df = df.loc[(df["year_x"] >= df["year_y"]) & (df["year_x"] - df["year_y"] <=3)]
index_df
```
